```python
def create_dataBase(url):
    response_hosp=requests.get(url)
    soup_hos=BeautifulSoup(response_hosp.content,"html.parser")
    tables = soup_hos.find_all("table")
    # Obtain information from tag <table>
    table1 = soup_hos.find_all('table')
    # Obtain every title of columns with tag <th>
    headers = []
    for i in table1[1].find_all('th'):
        title = i.text
        headers.append(title)
    #Creation du Dataframe
    df = pd.DataFrame(columns=headers)   
    # Create a for loop to fill df
    for j in table1[1].find_all('tr')[1:]:
        row_data = j.find_all('td')
        row = [i.text for i in row_data]
        length = len(df)
        df.loc[length] = row
    return df
    
    
    
```
