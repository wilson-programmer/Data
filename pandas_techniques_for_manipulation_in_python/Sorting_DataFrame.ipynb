# Here we will discuss about sorting dataFrame

```python
#import modules
import pandas as pd
import numpy as np
```

```python
# Import DataSet
df = pd.read_csv('sales_subset.csv')
df.head()
```

```python
# Drop the Unnamed: 0 column
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.sample(6)
```

```python
#Get information about the dataFrame
df.info()
```

```python
df
```

```python
# let see the department with the highest sale price
df.sort_values('weekly_sales')
```

```python
#sorting weekly_sales in descending order to see the department which make the biggest sale
df.sort_values('weekly_sales', ascending=False)
```

# we can see that the highest weekly_sales was during the working days


```python
#Sorting by multiple variables
df.sort_values(['weekly_sales','temperature_c'], ascending = [False,False])
```
